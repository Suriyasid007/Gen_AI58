```python
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.visualization import astropy_mpl_style

# 1. Locate the file
# sdss_access downloads files to a specific local path based on the parameters
# Usually: $HOME/sas/dr18/sky/photoObj/frames/301/1234/1/frame-r-001234-1-0011.fits.bz2
file_path = access.get_path('frame', run=1234, camcol=1, field=11, band='r')

# 2. Open the FITS file
# SDSS images are often compressed (.bz2), but astropy handles this automatically
with fits.open(file_path) as hdul:
    image_data = hdul[0].data  # The image pixels are in the first 'Extension'

# 3. Display the image
plt.style.use(astropy_mpl_style)
plt.figure(figsize=(10, 10))
plt.imshow(image_data, cmap='gray', origin='lower')

# Adding a colorbar to see intensity (photon counts)
plt.colorbar(label='Flux')
plt.title('SDSS Frame: Run 1234, Camcol 1, Field 11 (r-band)')
plt.show()
```
