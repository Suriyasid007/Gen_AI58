```python
import pytesseract
from PIL import Image
import os
from google.colab import files
from io import BytesIO # Import BytesIO for handling file content in memory

# 1. Upload your batch of invoices (can select multiple files at once)
print("Please select your English and Spanish invoice images:")
# Corrected: Use files.upload() to allow user to select multiple files
uploaded = files.upload()

# 2. Iterate through each uploaded file in the batch
for file_name, file_content in uploaded.items(): # Iterate over items to get both name and content
    print(f"\n--- Processing: {file_name} ---")

    try:
        # Open the image file from its content using BytesIO
        img = Image.open(BytesIO(file_content))

        # Extract text specifying BOTH English and Spanish
        # The engine will use both data sets to evaluate the characters
        text = pytesseract.image_to_string(img, lang='eng+spa')

        # Display the results
        print(text)

    except Exception as e:
        print(f"Error processing {file_name}: {e}")
```
